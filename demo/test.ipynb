```python
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', None)
import sys
sys.path.append('../')
import pathlib
directory_path = str(pathlib.Path('__notebook__').parent.resolve())


from tabularmagic.api import TabularMagic
from tabularmagic.api.mlR import (LinearR, KBestSelectorR)



df_train = pd.read_csv(f'{directory_path}/house_price_data/train.csv', index_col='Id')

display(df_train.head())
print('df shape:', df_train.shape)
```

```python
tm = TabularMagic(df_train, test_size=0.3)
tm
```

```python
report = tm.ml_regression(
    models=[
        LinearR(type='ols'),
        LinearR(type='l2')
    ],
    y_var='SalePrice',
    X_vars=['LotFrontage', 'LotArea', 'SaleCondition'],
    outer_cv=5
)
```

```python
report['LinearR(ols)'].train().fit_statistics()
```

```python
report['LinearR(ols)'].test().fit_statistics()
```
